Q2. Show the Bar Graphs Representing the Departure and Arrival Time

```python
df['departure_time'].value_counts()
```

```python
df['arrival_time'].value_counts()
```

```python
plt.figure(figsize=(16,4))
plt.subplot(1,2,1)
plt.title("Arrival Time Vs Arrival Frequency")
plt.bar(df['arrival_time'].value_counts().index,df['arrival_time'].value_counts().values, color=['r','b'])
plt.xlabel('Arrival Time')
plt.ylabel('Arrival Frequency')
plt.subplot(1,2,2)
plt.title("Departure Time Vs Departure Frequency")
plt.bar(df['departure_time'].value_counts().index,df['departure_time'].value_counts().values, color=['y','b'])
plt.xlabel('Departure Time')
plt.ylabel('Departure Frequency')
plt.show()
```

```python
plt.figure(figsize=(16,4))
plt.subplot(1,2,1)
plt.title("Departure Time Vs Arrival Time")
plt.bar(df['departure_time'].value_counts().index,df['arrival_time'].value_counts().index, color=['g','b'])
plt.xlabel('Departure Time')
plt.ylabel('Arrival Time')

plt.subplot(1,2,2)
plt.title("Arrival Time Vs Departure Time")
plt.bar(df['arrival_time'].value_counts().index,df['departure_time'].value_counts().index, color=['r','b'])
plt.xlabel('Arrival Time')
plt.ylabel('Departure Time')
plt.show()
```

Q5. Does the Ticket Price changes based on the Departure time and Arrival time

```python
df.groupby('departure_time')['price'].mean()
```

```python
df.groupby('arrival_time')['price'].mean()
```

```python
sns.catplot(x='departure_time',y='price',kind='bar',data=df,height=6,aspect=1.5)
plt.xticks(rotation=30)
plt.show()
```

```python
sns.catplot(x='arrival_time',y='price',kind='bar',data=df,height=6,aspect=1.5)
plt.xticks(rotation=30)
plt.show()
```

```python
sns.relplot(x='arrival_time',y='price',data=df, col='departure_time', kind='line')
plt.xticks(rotation=30)
plt.show()
```

```python
sns.relplot(x='arrival_time',y='price',data=df, col='departure_time', kind='scatter')
plt.xticks(rotation=30)
plt.show()
```
